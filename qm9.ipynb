```python
import rdkit.Chem as Chem
from rdkit.Chem import rdDetermineBonds

from bond_predictor import predict
```

```python
mol = Chem.MolFromXYZFile("../qm9-xyz/dsgdb9nsd_053848.xyz")
mol
```

```python
cloned_mol = Chem.Mol(mol)
rdDetermineBonds.DetermineBonds(cloned_mol)
print(cloned_mol.GetNumBonds(), cloned_mol.GetNumAtoms())
cloned_mol
```

```python
model = predict.load_model("/home/ameyad/bond-predictor/exported/bond_predictor.pt2")
mol_with_bonds = predict.predict_bonds(model, mol)
for bond in mol_with_bonds.GetBonds():
    print(bond.GetBeginAtomIdx(), bond.GetEndAtomIdx(), bond.GetBondType())

Chem.SanitizeMol(mol_with_bonds)
print(mol_with_bonds.GetNumBonds(), mol_with_bonds.GetNumAtoms())
mol_with_bonds
```
